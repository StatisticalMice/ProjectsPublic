```r
require(tibble)
require(dplyr)
require(microbenchmark)
require(zoo)
```

```r
N = 10^8
```

```r
df = tibble(data = rnorm(N)) %>%
    mutate(sma = zoo::rollmean(data, k = 10, fill = NA))
```

```r
df
```

```r
toc()
```

```r
res = microbenchmark(rnorm(N), times=5)
```

```r
print(res)
```

```r
res2 = microbenchmark(mutate(df, sma = lag(data, 10)), times=5)
```

```r
print(res2)
```

```r
res3 = microbenchmark(mutate(df, sma = zoo::rollmean(data, k = 10, fill = NA)), times=1)
```

```r
print(res3)
```
