```r
require(tibble)
require(dplyr)
require(microbenchmark)
```

```r
N = 10^8
```

```r
df = tibble(data = rnorm(N)) %>%
    mutate(sma = lag(data, 10))
```

```r
df
```

```r
toc()
```

```r
res = microbenchmark(rnorm(N), times=10)
```

```r
print(res)
```

```r
res2 = microbenchmark(mutate(df, sma = lag(data, 10)), times=10)
```

```r
print(res2)
```
